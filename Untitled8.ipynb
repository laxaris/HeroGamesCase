```python
import asyncio
from playwright.async_api import async_playwright
import nest_asyncio

# Mevcut event loop'u düzelt
nest_asyncio.apply()

async def scrape(base_url, start_page, end_page, output_file):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)  # Arka planda çalıştır
        page = await browser.new_page()

        total_articles = 0  # Toplam haber sayacı
        with open(output_file, "w", encoding="utf-8") as file:
            for page_num in range(start_page, end_page + 1):
                # URL'yi oluştur
                url = f"{base_url}?page={page_num}"
                print(f"Sayfa: {url} taranıyor...")

                # Sayfayı aç
                await page.goto(url)

                # Tüm haber başlıklarını ve linklerini çek
                articles = await page.locator("div[data-format-theme='0'] a").all()
                total_articles += len(articles)

                for article in articles:
                    title = await article.get_attribute("aria-label")  # Başlığı çek
                    link = await article.get_attribute("href")  # Linki çek

                    # Tam URL oluştur
                    if link and not link.startswith("http"):
                        link = f"https://www.theguardian.com{link}"

                    # Dosyaya yaz
                    file.write(f"'{link}',")

        # Tarayıcıyı kapat
        await browser.close()

        # Toplam haber sayısını yazdır
        print(f"Toplam {total_articles} haber bulundu.")
        print(f"Haberler '{output_file}' dosyasına kaydedildi.")

# Base URL ve sayfa aralığı
base_url = "https://www.theguardian.com/world/turkey"  # Sabit URL
start_page = 1  # Başlangıç sayfa numarası
end_page = 20  # Bitiş sayfa numarası
output_file = "tum_haberler.txt"  # Çıkış dosyasının adı


# Asenkron fonksiyonu çalıştır
await scrape(base_url, start_page, end_page, output_file)
```

```python
import asyncio
from playwright.async_api import async_playwright
import csv
import time


async def scrape_economist_updated():
    start_url = "https://www.economist.com/topics/turkey"
    pages_to_scrape = 30
    output_csv_file = "economist_articles.csv"
    articles = []

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        context = await browser.new_context()
        page = await context.new_page()

        # Navigate to the start URL
        await page.goto(start_url)
        for i in range(pages_to_scrape):
            print(f"Scraping page {i + 1}...")

            # Wait for the articles to load
            await page.wait_for_selector("h3.css-1jzypbl", timeout=60000)  # Wait for titles to be visible

            # Extract articles
            article_elements = await page.query_selector_all("div.css-0")
            for article in article_elements:
                # Extract title and description using the correct selectors
                title = await article.query_selector("h3.css-1jzypbl")
                description = await article.query_selector("p.css-llqkfq")

                title_text = await title.inner_text() if title else ""
                description_text = await description.inner_text() if description else ""

                # Append only if both title and description are present
                if title_text and description_text:
                    articles.append({"title": title_text, "description": description_text})

            # Click the "Next" button
            try:

              next_button = await page.wait_for_selector(
                  "a.ds-link-with-arrow[data-analytics='topic_page:next']",
                  timeout=120000  # Wait up to 60 seconds for the button
              )


              if await next_button.is_visible():
                  # Log the button's href attribute for debugging
                  next_href = await next_button.get_attribute("href")
                  await page.goto(next_href)


                  continue
              else:
                  print("'Next' button is not visible on the page.")
                  break
            except Exception as e:
                print(f"Error interacting with 'Next' button: {e}")
                break

        # Close the browser
        await browser.close()

        # Save results to a CSV file
        with open(output_csv_file, "w", newline="", encoding="utf-8") as csvfile:
            fieldnames = ["title", "description"]
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(articles)

        print(f"Scraping completed. Data saved to {output_csv_file}.")

# Run the scraper
asyncio.run(scrape_economist_updated())
```

```python
from playwright.async_api import async_playwright

async def scrape_lastampa_links():
    base_url = "https://www.lastampa.it/argomenti/turchia/"
    pages_to_scrape = 25  # Sayfa sayısı (1'den 9'a kadar)
    output_file = "lastampa_links.txt"
    all_links = []

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        for page_num in range(1, pages_to_scrape + 1):
            url = f"{base_url}{page_num}/"
            print(f"Scraping page: {url}")
            await page.goto(url)
            await page.wait_for_selector("h2.entry__title a")  # Updated selector for links

            # Makale linklerini topla
            links = await page.query_selector_all("h2.entry__title a")
            for link in links:
                href = await link.get_attribute("href")
                if href:  # Eğer link varsa
                    all_links.append(f'"{href}"')  # Tırnak içinde ekle

        # Tarayıcıyı kapat
        await browser.close()

        # Linkleri txt dosyasına kaydet
        with open(output_file, "w", encoding="utf-8") as f:
            f.write(",\n".join(all_links))  # Her URL'yi virgülle ayır ve yeni satıra ekle

        print(f"Scraping completed. Links saved to {output_file}.")

# Asenkron fonksiyonu çalıştır
import asyncio
asyncio.run(scrape_lastampa_links())
```

```python
import csv
from playwright.async_api import async_playwright
from playwright_stealth import stealth_async

async def scrape_corriere_articles(base_url, pages_to_scrape, output_csv):
    # Prepare the CSV file for output
    with open(output_csv, "w", newline="", encoding="utf-8") as csvfile:
        fieldnames = ["Title", "Description"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        async with async_playwright() as p:
            browser = await p.chromium.launch(headless=True)  # Open browser window for debugging
            context = await browser.new_context(
                user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"
            )
            page = await context.new_page()

            # Apply stealth mode
            await stealth_async(page)

            for page_num in range(1, pages_to_scrape + 1):
                url = f"{base_url}{page_num}/"
                print(f"Scraping: {url}")
                try:
                    await page.goto(url)
                    await page.wait_for_selector("div.bck-media-news", timeout=30000)  # Ensure articles are loaded

                    # Find all articles on the page
                    articles = await page.query_selector_all("div.bck-media-news")
                    for article in articles:
                        try:
                            # Extract title and description using query_selector
                            title_element = await article.query_selector("a.has-text-black")
                            description_element = await article.query_selector("a.has-text-grey-dove")

                            # Get text content
                            title = await title_element.inner_text() if title_element else "No Title"
                            description = await description_element.inner_text() if description_element else "No Description"

                            # Write the data to the CSV
                            writer.writerow({"Title": title.strip(), "Description": description.strip()})
                        except Exception as e:
                            print(f"Error extracting article: {e}")
                            writer.writerow({"Title": "Error", "Description": "Error"})
                except Exception as e:
                    print(f"Error scraping page {page_num}: {e}")

            # Close the browser
            await browser.close()

    print(f"Scraping completed. Data saved to {output_csv}.")

# Define base URL, number of pages, and output file
base_url = "https://www.corriere.it/argomenti/turchia/"
pages_to_scrape = 12  # Change this to the number of pages to scrape
output_csv = "corriere_articles.csv"

# Run the scraper
import asyncio
asyncio.run(scrape_corriere_articles(base_url, pages_to_scrape, output_csv))
```

```python
import asyncio
from playwright.async_api import async_playwright

async def extract_links(url):
    # List to store extracted links
    extracted_links = []

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        print(f"Navigating to {url}")
        await page.goto(url)

        # Wait for the elements to load
        await page.wait_for_selector("a.ui-story-headline")

        # Extract all <a> tags with the class `ui-story-headline`
        links = await page.query_selector_all("a.ui-story-headline")
        for link in links:
            href = await link.get_attribute("href")
            if href:
                extracted_links.append(href)

        await browser.close()

    return extracted_links

# Main function to run the scraper
async def main():
    url = "https://news.sky.com/topic/turkey-5859"  # Replace with your target URL
    links = await extract_links(url)

    # Print the extracted links
    for link in links:
        print(link)

    # Save to a text file if needed
    with open("extracted_links.txt", "w", encoding="utf-8") as f:
        for link in links:
            f.write(link + "\n")

    print("Links saved to extracted_links.txt")

# Run the script
asyncio.run(main())
```

```python
import asyncio
import csv
from playwright.async_api import async_playwright

async def scrape_titles_and_descriptions(link_file, output_csv):
    # Read links from the text file
    with open(link_file, "r", encoding="utf-8") as file:
        links = [line.strip() for line in file if line.strip()]

    # Prepare the CSV file for output
    with open(output_csv, "w", newline="", encoding="utf-8") as csvfile:
        fieldnames = ["Title", "Description"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        async with async_playwright() as p:
            browser = await p.chromium.launch(headless=True)
            context = await browser.new_context()

            for url in links:
                try:
                    page = await context.new_page()
                    print(f"Scraping: {url}")
                    await page.goto("https://news.sky.com"+url)

                    # Wait for the title and description elements
                    await page.wait_for_selector("span.sdc-article-header__long-title", timeout=30000)

                    # Extract the title and description
                    title = await page.locator("span.sdc-article-header__long-title").inner_text()
                    description = await page.locator("p.sdc-article-header__sub-title").inner_text()

                    # Write data to the CSV file
                    writer.writerow({

                        "Title": title.strip(),
                        "Description": description.strip()
                    })

                except Exception as e:
                    print(f"Error scraping {url}: {e}")
                    writer.writerow({
                        "Title": "Error",
                        "Description": "Error"
                    })

            await browser.close()

    print(f"Scraping completed. Data saved to {output_csv}.")

# Define input and output files
link_file = "skynews_links.txt"  # Input file containing URLs
output_csv = "skynews_scraped_data.csv"  # Output file for scraped data

# Run the scraper
asyncio.run(scrape_titles_and_descriptions(link_file, output_csv))
```

```python
import asyncio
import csv
from playwright.async_api import async_playwright

async def scrape_bbc_news(base_url, output_csv, max_pages=5):
    # Prepare the CSV file for output
    with open(output_csv, "w", newline="", encoding="utf-8") as csvfile:
        fieldnames = ["Title", "Description"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        async with async_playwright() as p:
            browser = await p.chromium.launch(headless=True)
            page = await browser.new_page()

            current_page = 1
            while current_page <= max_pages:
                print(f"Scraping page {current_page}...")
                try:
                    # Navigate to the page
                    await page.goto(base_url)
                    await page.wait_for_selector("div[data-testid='liverpool-card']")

                    # Extract titles and descriptions
                    articles = page.locator("div[data-testid='liverpool-card']")
                    count = await articles.count()
                    for i in range(count):
                        try:
                            title = await articles.nth(i).locator("h2[data-testid='card-headline']").inner_text()
                            description = await articles.nth(i).locator("p[data-testid='card-description']").inner_text()
                            writer.writerow({"Title": title.strip(), "Description": description.strip()})
                        except Exception as e:
                            print(f"Error extracting article {i} on page {current_page}: {e}")

                    # Check if the "Next" button exists and is enabled
                    next_button = page.locator("button[data-testid='pagination-next-button']")
                    if await next_button.is_enabled():
                        await next_button.click(timeout=12000000)
                        current_page += 1
                        await page.wait_for_timeout(2000)  # Small delay to ensure next page loads
                    else:
                        break
                except Exception as e:
                    print(f"Error scraping page {current_page}: {e}")
                    break

            await browser.close()

    print(f"Scraping completed. Data saved to {output_csv}.")

# Define base URL and output file
base_url = "https://www.bbc.com/news/topics/c207p54mdq3t"
output_csv = "bbc_news_data.csv"

# Run the scraper
asyncio.run(scrape_bbc_news(base_url, output_csv, max_pages=41))
```

```python
import asyncio
from playwright.async_api import async_playwright
import csv
import nest_asyncio
from google.colab import files  # Colab'den indirme işlemi için

# Mevcut event loop'u düzelt
nest_asyncio.apply()

async def scrape_lemond(base_url, output_file):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        # Sayfayı aç
        await page.goto(base_url)

        # CSV dosyasını oluştur
        with open(output_file, mode="w", newline="", encoding="utf-8") as file:
            writer = csv.writer(file)
            writer.writerow(["Başlık", "Açıklama"])  # Başlık satırı

            # Tüm başlık ve açıklamaları seç
            titles = await page.locator("h3.teaser__title").all()
            descriptions = await page.locator("p.teaser__desc").all()

            for title, desc in zip(titles, descriptions):
                # Başlık ve açıklamayı çek
                title_text = await title.text_content()
                desc_text = await desc.text_content()

                # CSV'ye yaz
                writer.writerow([title_text.strip(), desc_text.strip()])

        # Tarayıcıyı kapat
        await browser.close()

        print(f"Veriler '{output_file}' dosyasına kaydedildi.")

# URL ve çıkış dosyası
base_url = "https://www.lemonde.fr/turquie"  # Belirttiğin site
output_file = "lemond_data.csv"  # Çıkış dosyası adı

# Asenkron fonksiyonu çalıştır
await scrape_lemond(base_url, output_file)

# Google Colab'de dosyayı indir
print(f"'{output_file}' indiriliyor...")
files.download(output_file)
```

```python
import asyncio
from playwright.async_api import async_playwright
from playwright_stealth import stealth_async
import csv

def save_to_csv(data, output_file):
    with open(output_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Title", "Description"])
        writer.writerows(data)

async def scrape_foxnews(base_url, load_more_count, output_file):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)  # Set headless=False to debug
        context = await browser.new_context(
            user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        )
        page = await context.new_page()

        await stealth_async(page)  # Enable stealth mode

        all_data = []

        try:
            print(f"Scraping URL: {base_url}")
            await page.goto(base_url, timeout=60000)

            # Click "Show More" button multiple times
            for _ in range(load_more_count):
                try:
                    show_more_button = page.locator("div.button.load-more.js-load-more")
                    if await show_more_button.is_visible():
                        print("Clicking 'Show More' button...")
                        await show_more_button.hover()  # Hover over the button
                        await show_more_button.click(force=True)  # Force click in case of overlays
                        await page.wait_for_timeout(3000)  # Wait for new content to load
                    else:
                        print("No more 'Show More' button found.")
                        break
                except Exception as e:
                    print(f"Error clicking 'Show More': {e}")
                    break

            # Extract titles and descriptions
            titles = await page.locator("h4.title a").all_text_contents()
            descriptions = await page.locator("p.dek").all_text_contents()

            for title, description in zip(titles, descriptions):
                all_data.append([title.strip(), description.strip()])

            print(f"Scraping completed.")

        except Exception as e:
            print(f"Failed to scrape {base_url}: {e}")

        await browser.close()

        # Save results to CSV
        save_to_csv(all_data, output_file)
        print(f"Data saved to {output_file}")

# Parameters
base_url = "https://www.foxnews.com/category/world/world-regions/turkey"
load_more_count = 15  # Number of times to click "Show more"
output_file = "foxnews_data.csv"

# Run the scraper
asyncio.run(scrape_foxnews(base_url, load_more_count, output_file))
```

```python

import asyncio
from playwright.async_api import async_playwright
import csv
import nest_asyncio

# Colab için event loop sorununu çözmek
nest_asyncio.apply()

# Base URL for Le Figaro's "Turquie" tag pages
base_url = "https://www.lefigaro.fr/tag/turquie?page="

# Number of pages to scrape
start_page = 1
end_page = 8  # Adjust this as needed for more pages

# Initialize a list to store the data
scraped_data = []

async def scrape_with_playwright():
    async with async_playwright() as p:
        # Launch browser
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        for page_num in range(start_page, end_page + 1):
            # Construct the page URL
            url = f"{base_url}{page_num}"
            # Navigate to the page
            await page.goto(url)

            # Wait for the content to load
            await asyncio.sleep(3)  # Adjust timeout if necessary

            # Extract all article containers
            articles = await page.locator('article').all()
            print(articles)

            # Loop through each article
            for article in articles:
                # Extract the title
                title = await article.locator('h2.fig-ranking-profile-headline').inner_text(timeout=1000)
                # Extract the description if present
                try:
                    description = await article.locator('p.fig-ranking-profile-standfirst').inner_text(timeout=1000)
                except:
                    description = None

                # Skip articles without a description
                if not description:
                    continue

                # Append the data
                scraped_data.append({"Title": title, "Description": description})

        # Close the browser
        await browser.close()

# Run the asyncio function
await scrape_with_playwright()

# Save the data to a CSV file
output_path = 'lefigaro_scraped_data.csv'
with open(output_path, 'w', encoding='utf-8', newline='') as csvfile:
    csvwriter = csv.DictWriter(csvfile, fieldnames=["Title", "Description"])
    csvwriter.writeheader()
    csvwriter.writerows(scraped_data)

# Print the path to the saved CSV file
print(f"Scraping completed! CSV saved at: {output_path}")
```

```python
import asyncio
from playwright.async_api import async_playwright
from playwright_stealth import stealth_async
import nest_asyncio
from google.colab import files  # Colab için

# Mevcut event loop'u düzelt
nest_asyncio.apply()

async def scrape_with_stealth_and_screenshot(base_url, start_page, end_page, output_file, screenshot_dir):
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)  # Gösterimli modda başlat
        context = await browser.new_context(
            user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        )
        page = await context.new_page()

        await stealth_async(page)  # Stealth modunu etkinleştir

        with open(output_file, mode="w", encoding="utf-8") as file:
            for page_num in range(start_page, end_page + 1):
                url = f"{base_url}/p{page_num}/"
                print(f"Sayfa işleniyor: {url}")
                try:
                    # Sayfayı aç
                    await page.goto(url, wait_until="load")

                    # Ekran görüntüsü al
                    screenshot_path = f"{screenshot_dir}/page_{page_num}.png"
                    await page.screenshot(path=screenshot_path, full_page=True)
                    print(f"Ekran görüntüsü kaydedildi: {screenshot_path}")

                    # HTML içeriği al
                    html = await page.content()
                    file.write(f"\n--- Sayfa: {url} ---\n")
                    file.write(html + "\n")
                except Exception as e:
                    print(f"Hata oluştu: {url} - {e}")
                    continue

        # Tarayıcıyı kapat
        await browser.close()

        print(f"Tüm HTML verileri '{output_file}' dosyasına kaydedildi.")

# Base URL ve sayfa aralığı
base_url = "https://www.spiegel.de/thema/tuerkei"  # Sabit URL
start_page = 1  # Başlangıç sayfa numarası
end_page = 3  # Bitiş sayfa numarası
output_file = "spiegel_raw_data.txt"  # HTML verilerinin kaydedileceği dosya
screenshot_dir = "screenshots"  # Ekran görüntülerinin kaydedileceği klasör

# Asenkron fonksiyonu çalıştır
await scrape_with_stealth_and_screenshot(base_url, start_page, end_page, output_file, screenshot_dir)

# Google Colab'den dosyayı indir
print(f"'{output_file}' indiriliyor...")
files.download(output_file)
```
